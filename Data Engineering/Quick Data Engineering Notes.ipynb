## Data Engineering

### Types of data (easiest to work with to most difficult)

1. Structured Data - usually comes from relational databases
2. Semi-structured Data - XML, CSV or JSON form. Different file types to store data  [kaggle](kaggle.com) has free data sets.
3. Unstructured Data - email, pdfs
4. Binary Data - audio & image files

### What does a Data Engineer do?
* Data Mining - preprocessing and extracting some knowledge from the data
* Big Data - we have a lot of data and a lot of variables. So much that you can't run it on a laptop. Usually, it requires multiple computers or in a cloud.
* Data Pipeline - A pipeline that a data engineer builds to allow us to flow data from unknown data to a more useful form.

### Data Flow
1. Aquire data from different sources and the raw data goes into a **data lake**
2. From the **data lake** we do data transformations into a **data warehouse**
3. The **data warehouse** is where we store data that is useful to the business
4. The business then can query the **data warehouse** since it's now structured specifically for insights.

### Data Engineer tasks
1. Build ETL pipelines - **extract** data, **transform** it and **load** it into a warehouse.
2. Builds tools that are used for analyzing data
3. Maintain the data warehouse and data lakes

### Main 3 types of Databases
1. Relational Database - Allows you to use SQL to make transactions. Good for ACID transactions.
    * **ACID Transactions**
        * **Atomicity** ensures that all operations within a transaction are treated as a single unit, which all succeed or fail together.
        * **Consistency** ensures that a transaction can only bring the database from one valid state to another, allowing only valid data to be written and preventing data corruption. This stability is maintained whether the transaction succeeds or fails.
        * **Isolation** determines how and when changes made by one transaction become visible to others.
            * If multiple transactions occur at the same time, they will not impact otherâ€™s executions. Complete isolation of concurrently running transactions could mean that one transaction cannot perform an insert into a table being queried by another transaction. Therefore, real-world considerations usually require a compromise between perfect transaction isolation and performance.
        * **Durability** ensures that transaction results are permanently stored in the system (such as backups and transaction logs). Modifications must persist even if the system fails or there is a loss of power.

2. NoSQL - Key based storage, like a dictionary. A little more disorganized, but allows us to store a lot more data across different machines.
3. NewSQL - New type that tries to solve the problems of the relational database size, while having the flexability of a NoSQL database

### Hadoop and MapReduce
* Hadoop is a data lake solution that became popular as big data began to take off
* It's popularity is a result of two things
    1. HDFS - Hadoop Distributed File System. This allowed it to store data on multiple computers making it very scalable
    2. MapReduce - Allowed us to perform batch processing jobs on multiple machines
* **Hive** makes it feel like your HDFS is on a single computer

### Apache Spark & Apache Flink
* Spark improved the MapReduce of Hadoop, but changed it to **in memory** processing.
    * It is the go to batch processing software
    * Spark will run ETL jobs
* Real time processing begame to replace overnight batch jobs.
* Flink took over the real-time stream processing

### Kafka & Stream Processing
* Kafka is a distributed streaming platform.
    * Allows us to receive messages and pass them on to different places.
    * Data comes in from different sources and then it gets passed to Hadoop, Spark, AWS S3, or other Databases
