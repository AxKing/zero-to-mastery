### Anatomy of a matplotlib figure

```python
# Matplotlib exmaple workflow
# 0. import matplot lib
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare Data
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="simple plot",
      xlabel = "x-axis",
      ylabel="ylabel")

# 5. Save and show
fig.savefig("../data/images/sample-plot.png")
plt.show()
```

#### Scatter Plot and Bar Plot
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(1,10,100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter":10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut butter Store",
      ylabel = "Price ($)");
```

#### Histograms and subplots

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data
x = np.random.randn(1000)
```

```python
# Make some data for histograms and plot it
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

### Subplots Option 2

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# plot to each different different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(10000));
```
