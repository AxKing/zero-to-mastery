### Anatomy of a matplotlib figure

```python
# Matplotlib exmaple workflow
# 0. import matplot lib
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare Data
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="simple plot",
      xlabel = "x-axis",
      ylabel="ylabel")

# 5. Save and show
fig.savefig("../data/images/sample-plot.png")
plt.show()
```

## Plotting from Pandas DataFrames 5

### which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
print(len(over_50), len(heart_disease))
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

## Plotting Data Frames Part 6

```python
# OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Create a subplot using thalach
fig, ax = plt.subplots()

scatter = ax.scatter(x=over_50['chol'],
                    y=over_50['thalach'],
                    c=over_50['target'])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel = "Cholesterol",
      ylabel="Blood Pressure");
```

### Plotting from Pandas DataFrames 7

```python
# Subplot of chol, age, Thalach
fig , (ax0, ax1) = plt.subplots(nrows=2, ncols = 1, figsize= (10,10), sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Add the labels 
ax0.set(title="Heart Disease and Cholestoral Levels",
       ylabel="Cholestoral")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thal"],
                     c=over_50["target"])

# Add the labels
ax1.set(title="Max Heart Rate and Heart Disease",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a line at the mean thal
ax1.axhline(y=over_50['thal'].mean(),
           linestyle="--");

# Add a title to the bigger figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

### Customizing plots part 2

```python
# Set the Style
plt.style.use('seaborn-whitegrid')


fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50["target"],
                    cmap='winter') # changes the color scheme

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
# Customizing the x and y axis limitations

# Subplot of chol, age, Thalach
fig , (ax0, ax1) = plt.subplots(nrows=2, ncols = 1, figsize= (10,10), sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter')
# Customize ax0
ax0.set(title="Heart Disease and Cholestoral Levels",
       ylabel="Cholestorol")
# Changes the x axis limits
ax0.set_xlim([50,80])
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap='winter')

# Add the labels
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
# Changes the x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a line at the mean thal
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle="--");

# Add a title to the bigger figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
